```python
def fit(w):
    w0 = w[:66].reshape((11, 6))
    w1 = w[66:72].reshape((6,1))
    b0 = np.array([0.,0.,0.,0.,0.,0.])
    b1 = np.array([0.])
    model.layers[0].set_weights([w0, b0])
    model.layers[1].set_weights([w1, b1])
    model.compile(optimizer="SGD", loss="binary_crossentropy")
    loss = model.evaluate(x_train, y_train, verbose=0)
    return -loss
```

# Genetic Algorithm

```python
def genetic_algorithm(pop_size, gens, mut_rate):
    pop = [np.random.uniform(low=-1, high=1, size=77) for _ in range(pop_size)]   
    for gen in range(gens):
        fitness = [fit(x) for x in pop]
        idx = np.argsort(fitness)[-2:]
        parents = [pop[i] for i in idx]
        offsprings = []
        for _ in range(pop_size - len(parents)):
            p1, p2 = parents[0].flatten(), parents[1].flatten()

            cp = np.random.randint(0, len(p1))
            child = np.concatenate([p1[:cp], p2[cp:]])
            child = child.reshape(parents[0].shape)
 
            for i in range(len(child)):
                if np.random.uniform() < mut_rate:
                    child[i] += np.random.normal(loc=0, scale=0.1)
            offsprings.append(child)
        pop = parents + offsprings
        if (gen % 10 == 0):
            print("Gen:", gen)
            
    fitness = [fit(x) for x in pop]
    best_idx = np.argmax(fitness)
    best_w = pop[best_idx]
    return best_w
```

```python
best_weights = genetic_algorithm(pop_size=20, gens=50, mut_rate=0.1)
w0 = best_weights[:66].reshape((11, 6))
w1 = best_weights[66:72].reshape((6,1))
b0= np.array([0.,0.,0.,0.,0.,0.])
b1 = np.array([0.])

model.layers[0].set_weights([w0, b0])
model.layers[1].set_weights([w1, b1])
```

```python
model.compile(optimizer="SGD", loss="binary_crossentropy",metrics=['accuracy'])
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
accuracy_score(y_pred,y_test)
```
